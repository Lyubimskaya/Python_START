```python
from sympy.plotting import plot

from sympy import *
```

```python
x = Symbol("x")
f = x**3 - 50*x
f
```

```python
g = -x**4 + 88*x**2 - 241
g
h = f - g
```

```python
k = solve(h)
[[xelem.evalf(chop=True).round(2),f.subs(x,xelem).evalf(chop=True).round(2)] for xelem in k]
```

```python
plot(f,g)
```

```python
plot(h)
```
